```python
data['age'] = [w2n.word_to_num(item) if isinstance(item,str) else item for item in data['age']]
```

```python
data1=data.copy()
```

```python
data1.dropna(inplace=True)
data1.isnull().sum()
```

```python
data2 = data.copy()
mean_age=data2['age'].mean()
mode_age=data2['age'].mode()
mean_fare=data2['fare'].mean()
```

```python
#what is the most aplicable to fill in with?
# data2["age"]=data2["age"].fillna(0)
data2["age"]=data2["age"].fillna(mean_age)
data2["fare"]=data2["fare"].fillna(mean_fare)
```

```python
data2.isnull().sum()
```

```python
data3= data2.drop(['sex'],axis = 1)
```

```python
dummy = pd.get_dummies(data2["sex"], prefix="gender")
data3=pd.concat([data3,dummy],axis=1)
data3
```
